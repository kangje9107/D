```python
from selenium import webdriver
import time
import pandas as pd

#driver = webdriver.Chrome('../chromedriver')
driver = webdriver.Chrome(executable_path=r'C:\Users\20200329\Desktop\분석\chromedriver/chromedriver.exe')
driver.get('https://play.google.com/store/apps/details?id=com.mobiletalk.valley')


```

```python
SCROLL_PAUSE_TIME = 1.5
last_height = driver.execute_script("return document.body.scrollHeight")
num = 0
while True:
    # (1) 4번의  스크롤링
    for i in range(4):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(SCROLL_PAUSE_TIME)

    # (2) 더보기 클릭
    
    try : 
        driver.find_element_by_xpath("//span[@class='RveJvd snByac']").click()
    except  :
        print('reach the bottom of page')
        #break
    num += 1     
    # (3) 종료 조건
    
    new_height = driver.execute_script("return document.body.scrollHeight")
    print("[", num, "]", "last : " , last_height, "new : ", new_height)
    
    if new_height == last_height:
        break
    last_height = new_height
```

```python
#긴 리뷰가 있으면 '전체리뷰' 버튼 클릭
spread_review = driver.find_elements_by_xpath("//button[@jsaction='click:TiglPc']")
spread_review

for i in range(len(spread_review)):
    isTrue = spread_review[i].is_displayed()    # 보이는 것인지를 확인
    print("Element is visible? " + str(isTrue))
    if isTrue:
        spread_review[i].click()
        print(str(i)+"th more button is clicked and wait 2 secs...")
        time.sleep(2)
```

```python
reviews = driver.find_elements_by_xpath("//div[@jsname='fk8dgd']/span[(@jsname='bN97Pc')]")

for i in range(len(reviews)):
  print(str(i) + "\t" + reviews[i].text)
    
long_reviews = driver.find_elements_by_xpath("//div[@class='UD7Dzf']/span[@jsname='fbQN7e']")
#for i in range(len(long_reviews)):
#   print(str(i) + "\t" + long_reviews[i].text)

empty = [i for i,t in enumerate(reviews) if t.text=='']
long = [i for i,t in enumerate(reviews) if t.text!='']
empty == long    # True
print(long)
#merge two list



merged_review = [t.text if t.text!='' else long_reviews[i].text for i,t in enumerate(reviews)]    
```

```python
res_dict = []
for i in range(len(merged_review)):
    res_dict.append({
        'DATE' : dates[i].text,
        'STAR' : stars[i].get_attribute('aria-label'),
        'LIKE' : likes[i].text,
        'WRITER': writer_names[i].text,
        'REVIEW' : merged_review[i]
    })
```

```python
res_df = pd.DataFrame(res_dict)
res_df
```

```python
from bs4 import BeautifulSoup
html = driver.page_source
```

```python
bsObj = BeautifulSoup(html,'lxml')
bs4_allreviews = bsObj.find_all("div",{"class":"d15Mdf bAhLNe"})
print(len(bs4_allreviews))
```

```python
print(bs4_allreviews)
```

```python
from bs4.element import NavigableString
bs_result=[]
for i in bs4_allreviews :
    #print(i.find("div",{"class":'QoPmEb'}))
    #[t.text if t.text!='' else long_reviews[i].text for i,t in enumerate(reviews)]
    
    if i.select('div[class="LVQB0b"]') : 
    
        for j in i.select('div[class="LVQB0b"]'):
            li_j = list(j.stripped_strings)
            comment_writer = li_j[0]
            comment_date =li_j[1]
            comment_contents = li_j[2].replace("\n","")
            
    
    else: 
            comment_writer = ''
            comment_date =''
            comment_contents = ''
    
    bs_result.append({
        'CMT_DATE' : comment_date,
        'CMT_WRITER' : comment_writer,
        'COMMENT' : comment_contents
    })
        
    del comment_writer, comment_date, comment_contents
        
```

```python
print(len(bs_result))
for i in bs_result: 
    print(i)
```

```python
df_comments = pd.DataFrame(bs_result)
concat_df= pd.concat([res_df,df_comments], axis=1)
```
